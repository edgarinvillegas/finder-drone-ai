```python
#%env CUDA_VISIBLE_DEVICES=2

import torch, torchvision
from torchvision import datasets, models, transforms
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import time
import cv2
from torchsummary import summary

import numpy as np
import matplotlib.pyplot as plt
import os

from PIL import Image
```

```python
dataset_dir = 'dataset_turtle/turtle-img-square/'
```

```python
# Applying Transforms to the Data
augment_transforms = transforms.Compose([
    transforms.RandomResizedCrop(size=256, scale=(0.8, 1.0)),
    #transforms.RandomRotation(degrees=180),
    transforms.RandomHorizontalFlip(),
    transforms.RandomVerticalFlip(),
])
```

```python
data = datasets.ImageFolder(root=dataset_dir, transform=augment_transforms)
```

```python
len(data)
```

```python
for image_tuple in data:    
    image = image_tuple[0]
    plt.imshow(image)
    break            
```

```python
total_images = 2600
loops = total_images // len(data)
for i in range(loops):
    print(i)
    for idx, image_tuple in enumerate(data):    
        image = image_tuple[0]
        npimage = np.array(image)
        npimage = cv2.cvtColor(npimage, cv2.COLOR_BGR2RGB)
        r = np.random.rand()
        if r < 0.6:
            set = 'train'
        elif r < 0.8:
            set = 'valid'
        else:
            set = 'test'
        # for example: dataset/train/juana1.jpg
        cv2.imwrite('dataset_turtle/{}/turtle/{}_{}.jpg'.format(set, i, idx), npimage)
        #plt.imshow(image)
        #break
    #break
print('done')    
    
```
